```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
from datetime import datetime, timedelta
```

```python
produtos = ['Camiseta', 'Calça', 'Tênis', 'Meia', 'Roupa Intíma']
pesos_produtos = [0.4, 0.3, 0.15, 0.05, 0.1]
preco_produto = [50, 80, 150, 20, 20]

categorias = ['Marca A', 'Marca B', 'Marca C']
pesos_categorias = [0.6, 0.3, 0.10]
preco_cat = [15, 35, 5]

quantidades = [1, 2, 3, 4, 5, 6, 7, 8]
pesos_quantidade = [0.25, 0.3, 0.2, 0.1, 0.025, 0.025, 0.025, 0.025]

inicio = datetime(2023, 1, 1)
fim = datetime(2023, 12, 31)
```

```python
colunas = ['ID', 'Data', 'Produto', 'Categoria', 'Quantidade', 'Preço']
vendas = pd.DataFrame(columns=colunas)

for i in range(500):
  id = i + 1

  id_prod = [0, 1, 2, 3, 4]
  id_cat = [0, 1, 2]

  ip = random.choices(id_prod, weights=pesos_produtos, k=1)[0]
  ic = random.choices(id_cat, weights=pesos_categorias, k=1)[0]
  quant = random.choices(quantidades, weights=pesos_quantidade, k=1)[0]

  prod = produtos[ip]
  cat = categorias[ic]

  preco_medio = preco_produto[ip] + preco_cat[ic]
  preco = np.random.normal(preco_medio, 10)
  preco =  round(preco / 5) * 5

  diaNum = random.randint(0, 365)

  dia = inicio + pd.DateOffset(days=diaNum)

  observacao = pd.DataFrame({'ID': [id], 'Data': [dia], 'Produto': [prod], 'Categoria': [cat], 'Quantidade': [quant], 'Preço': [preco]})
  vendas = pd.concat([vendas, observacao], ignore_index=True)
```
